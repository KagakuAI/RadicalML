```python
import os
import pandas as pd

from pathlib import Path

from sklearn.model_selection import train_test_split

from qsarlazy import LazyML, LazyQSARmil
```

### Load data

```python
# df_train = pd.read_csv("qc_data/train.csv", header=None)
# df_test = pd.read_csv("qc_data/test.csv", header=None)

df_train = pd.read_csv("../qsarcons/bench_data/ace/CHEMBL1862_Ki/train.csv", header=None)
df_test = pd.read_csv("../qsarcons/bench_data/ace/CHEMBL1862_Ki/test.csv", header=None)

df_train, df_val = train_test_split(df_train, test_size=0.2, random_state=42)
```

### 1. LazyML

```python
output_folder = Path("result_basic")

model = LazyML(task="regression", output_folder=output_folder, n_cpu=30)
model.run(df_train, df_val, df_test)
```

```python
df_val_basic = pd.read_csv(os.path.join(output_folder, "val.csv"))
df_test_basic = pd.read_csv(os.path.join(output_folder, "test.csv"))
df_test_basic
```

### 2. QSARmil

```python
output_folder = Path("result_mil")

model = LazyQSARmil(task="regression", output_folder=output_folder, n_cpu=30)
model.run(df_train, df_val, df_test)
```

```python
df_val_mil = pd.read_csv(os.path.join(output_folder, "val.csv"))
df_test_mil = pd.read_csv(os.path.join(output_folder, "test.csv"))
df_test_mil
```

### 3. QSARcons

```python
from qsarcons.consensus import RandomSearchRegressor, SystematicSearchRegressor, GeneticSearchRegressor
```

```python
df_val_cons = df_val_mil[["SMILES", "Y_TRUE"]]
df_test_cons = df_test_mil[["SMILES", "Y_TRUE"]]
```

```python
cons_size = "auto"
metric = "rmse"

method_list = [
               (RandomSearchRegressor(cons_size=cons_size, n_iter=1000, metric=metric), 'Random'),
               (SystematicSearchRegressor(cons_size=cons_size, metric=metric), 'Systematic'),
               (GeneticSearchRegressor(cons_size=cons_size, n_iter=50, pop_size=100, mut_prob=0.2, metric=metric), 'Genetic'),
               ]
```

```python
# BasicML consensus
x_val, y_val = df_val_basic.iloc[:, 2:], df_val_basic.iloc[:, 1]
x_test, y_test = df_test_basic.iloc[:, 2:], df_test_basic.iloc[:, 1]
for method_func, method_name in method_list:
    cons = method_func.run(x_val, y_val)
    df_val_cons[f"QSARcons|{method_name}|BasicML"] = method_func._consensus_predict(x_val[cons])
    df_test_cons[f"QSARcons|{method_name}|BasicML"] = method_func._consensus_predict(x_test[cons])

# QSARmil consensus
x_val, y_val = df_val_mil.iloc[:, 2:], df_val_mil.iloc[:, 1]
x_test, y_test = df_test_mil.iloc[:, 2:], df_test_mil.iloc[:, 1]
for method_func, method_name in method_list:
    cons = method_func.run(x_val, y_val)
    df_val_cons[f"QSARcons|{method_name}|QSARmil"] = method_func._consensus_predict(x_val[cons])
    df_test_cons[f"QSARcons|{method_name}|QSARmil"] = method_func._consensus_predict(x_test[cons])
```

### Results

```python
from sklearn.metrics import r2_score
```

```python
final_test = pd.concat([
                        df_test_basic,
                        df_test_mil.iloc[:, 2:],
                        df_test_cons.iloc[:, 2:],
                       ], axis=1)
res = pd.DataFrame()
for model in final_test.iloc[:, 2:]:
    res.loc[model, "ACC"] = r2_score(final_test["Y_TRUE"], final_test[model])
```
